<p>Para determinar valores atipicos se calcularan el limite superior e inferior mediante los rangos intercuartilicos.No se tendran en cuenta los atributos de latitud y longitud

```python
df_describe=df_3er_filtrado.describe().T
df_describe
df_describe['RIC']=df_describe['75%']-df_describe['25%']
df_describe['lim_inf']=df_describe['25%']-1.5*df_describe['RIC']
df_describe['lim_sup']=df_describe['75%']+1.5*df_describe['RIC']
df_describe.T
```

Veamos cuantos datos estan por fuera del limite superior, el limite inferior no tiene sentido revisarlo puesto que da valores negativos

```python
cant_instancias=[]
cant_excluidos=[]
for i in df_describe.index:
    cant_instancias.append(df_3er_filtrado[i].notnull().sum())
    cant_excluidos.append(df_3er_filtrado[df_3er_filtrado[i]>df_describe.loc[i,'lim_sup']].shape[0])
atipicos={'lim_sup': df_describe.loc[:,'lim_sup'], 'cant_instancias':cant_instancias , 'cant_excluidos':cant_excluidos}
dfatipicos = pd.DataFrame(data=atipicos)
dfatipicos['porc_excluidos']=100*(dfatipicos['cant_excluidos'])/dfatipicos['cant_instancias']
dfatipicos
```
